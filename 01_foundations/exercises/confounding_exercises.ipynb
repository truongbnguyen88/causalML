````python
# Helper function
def draw_dag(edges: List[Tuple[str, str]], title: str = "DAG", 
             node_size: int = 3000, font_size: int = 11, 
             figsize: Tuple[int, int] = (10, 6)):
    G = nx.DiGraph()
    G.add_edges_from(edges)
    
    plt.figure(figsize=figsize)
    pos = nx.spring_layout(G, k=2, iterations=50, seed=42)
    
    nx.draw(G, pos, 
            with_labels=True,
            node_color='lightblue',
            node_size=node_size,
            font_size=font_size,
            font_weight='bold',
            arrows=True,
            arrowsize=20,
            arrowstyle='->',
            edge_color='gray',
            width=2,
            connectionstyle='arc3,rad=0.1')
    
    plt.title(title, fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return G
````

## Exercise 5: Mediator vs Confounder (Medium)

**Learning Goal**: Practice distinguishing mediators from confounders.

**Task**:

   - Total effect (not controlling for mediator)
   - Direct effect (controlling for mediator)

**Example scenario**: Education → Job Skills → Income
- Someone might think Skills is a confounder ("smart people get skills AND education")
- But actually: Education causes Skills, which causes Income (mediator!)

<details>
<summary>Hint 1</summary>
Good scenarios for confusion:
- Treatment improves something, which then improves outcome
- The mediator looks like it could cause treatment (but doesn't)

Examples:
- Marketing → Brand Awareness → Sales
- Exercise → Fitness Level → Health
- Medication → Symptom Relief → Quality of Life
</details>

<details>
<summary>Hint 2</summary>
Simulation structure for mediator:
```python
treatment = np.random.binomial(1, 0.5, n)
mediator = a + b * treatment + noise  # Treatment CAUSES mediator
outcome = c + d * mediator + noise     # Mediator CAUSES outcome (no direct treatment effect)
```
Total effect = b * d
Direct effect ≈ 0
</details>

<details>
<summary>Hint 3</summary>
Compare:
- Model 1: Y ~ X (total effect = b * d)
- Model 2: Y ~ X + M (direct effect ≈ 0, M effect = d)

If you incorrectly control for M (thinking it's a confounder), you block the pathway and conclude no effect!
</details>
